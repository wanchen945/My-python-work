```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 模擬版資料做SVM

```python
x=np.array([[6,6],[10,8],[3,7],[8,7]])
```

```python
x
```

```python
x[:,0]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
y=np.array([1,2,1,2])
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=100)
```

## 打開函數學習機

```python
from sklearn.svm import SVC
```

```python
clf=SVC(gamma="auto")
```

## 學習

```python
clf.fit(x,y)
```

## 預測

```python
y_predict=clf.predict(x)
```

```python
y_predict
```

```python
y
```

```python
xr=np.random.rand(100)*7+10
yr=np.random.rand(100)*2+6
```

```python
plt.scatter(xr,yr)
```

```python
xr
```

```python
print("X",xr[:5])
print("Y",yr[:5])
```

```python
X=np.c_[xr,yr]
```

```python
Y=clf.predict(X)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap="Paired",s=150)
plt.scatter(X[:,0],X[:,1],c=Y,cmap="Paired",s=150)
```

## 結果呈現

```python
x_list=np.linspace(3,10,50)
y_list=np.linspace(6,8,50)
X,Y=np.meshgrid(x_list,y_list)
P=np.c_[X.ravel(),Y.ravel()]
C=clf.predict(P)
plt.scatter(P[:,0],P[:,1],c=C)
```

```python
X.ravel()
```

## 插播:CONTOUR用法

```python
xL=np.array([1,2,3,4])
yL=np.array([5,6,7,8])
X,Y=np.meshgrid(xL,yL)
```

```python
X
```

```python
Y
```

```python
Z=np.random.randint(1,3,(4,4))
plt.contour(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```

```python
Z=np.random.randint(1,3,(4,4))
plt.contourf(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```

## 資料視覺版終極化

```python
x_arr=np.arange(3,10,0.02)
y_arr=np.arange(6,8,0.02)
X,Y=np.meshgrid(x_arr,y_arr)
P=np.c_[X.ravel(),Y.ravel()]
z=clf.predict(P)
Z=z.reshape(X.shape)
plt.contourf(X,Y,Z,cmap="Paired",alpha=0.3)
plt.scatter(x[:,0],x[:,1],c=y,cmap="Paired",s=150)
```
